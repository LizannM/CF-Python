# 02. Sort into Price Ranges - User Defined

```python
# Define a function
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-Range Product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-Range Product'
    elif row['prices'] > 15:
        return 'High Range'
    else: return 'Not Enough Data'
```

```python
# Apply the funtion
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 03. Sort into Price Ranges - loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High Range Product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid Range Produt'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low Range Product'
```

# 05. Traffic using For-Loops

```python
df_ords_prods.head()
```

```python
df_ords_prods['orders_day_of_week'].value_counts(dropna=False)
```

```python
# loop to define traffic on each day of the week
result = []

for value in df_ords_prods["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append("Least Busy Day")
    else:
        result.append("Regularly Busy Day")
```

# 06. Busiest Days

New column to assign the two busiest days (0,1) and the two slowest days(3,4), the with the remaining being the regular days (2,5,6)

```python
# loop to define traffic on each day of the week
result = []

for value in df_ords_prods["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest Days")
    elif value == 4 or value == 3:
        result.append("Slowest Days")
    else:
        result.append("Regular Days")
```

```python
result
```

```python
df_ords_prods['Busiest_Days'] = result
```

```python
df_ords_prods['Busiest_Days'].value_counts(dropna=False)
```

# 07. New column for Busiest Time of Day

Time periods to be labeled as "Most Orders", "Average Orders" and "Fewest Orders"

```python
# Check frequency for each hour
df_ords_prods['order_time'].value_counts(dropna=False)
```

Most Orders (10,11,14,15,13,12,16,9)
Average Orders (17,8,18,19,20,7,21,22)
Fewest Orders (23,6,0,1,5,2,4,3)

```python
result = []

for value in df_ords_prods["order_time"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result.append("Most Orders")
    elif value == 23 or value == 6 or value == 0 or value == 1 or value == 5 or value == 2 or value == 4 or value == 3:
        result.append("Fewest Orders")
    else:
        result.append("Average Orders")
```

```python
result
```

```python
df_ords_prods['Busiest_Hours'] = result
```

```python
df_ords_prods['Busiest_Hours'].value_counts(dropna=False)
```
